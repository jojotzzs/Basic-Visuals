### Introductory Seaborn Visuals
##### Purpose: To build basic seaborn visuals using in house datasets

```python
## import libraries.

import pandas as pd
import numpy as mp
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
## load dataset from seaborn library

df = sns.load_dataset("tips")
df.head()
```

#### Swarmplots

```python
# visualise tip amount aganist the day of the week.
sns.swarmplot(x="day", y="tip",data=df)
sns.set_style("whitegrid")
plt.show()
```

```python
# visualise just the tip amounts.
sns.swarmplot(x=df["tip"])
sns.set_style("darkgrid")
plt.show()
```

```python
# visualise tip amounts by day but with gender by distinct colours.
gender_palette = ["#A833FF", "#FFAF33"]
sns.swarmplot(x="day", y="tip", hue="sex", palette= gender_palette, data=df)
plt.show()
```

#### Violin Plots

```python
# Visualise tips with the time of day by gender

sns.violinplot(x="time", y="tip",hue="sex",palette="dark", data=df, split=True)
plt.legend()
plt.show()
```
